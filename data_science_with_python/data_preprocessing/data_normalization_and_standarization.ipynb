**Feature Scaling**

Feature Scaling is a crucial step in machine learning to build some effective models. There are many aspects of Feature scaling which involves standardization, normalization, scaling of data to make useful models. By using help of these procedures we can make our model more precise more useful and efficient. It also helps in managing outliers present in data. This techniques gives an important glance at our models and help train them effectively and precisely by learning from the data.

**Why Using Feature Scaling**

> Feature scaling is crucial for machine learning models. many machine learning modelrely on the distances between data points or make assumptions about the data distribution. 

* here's why feature scaling become important :

    1. Prevent Features Domination : Feature scaling ensures that all features are treated equally.

    2. Distance based Algorithm : K-Nearest Neighbors (KNN), Support Vector Machines (SVM), and K-Means clustering use Euclidean distance

    3. Improved Convergence for Gradient-Based Models : Algorithms that optimize models through gradient descent methods, such as linear regression or neural networks, will converge faster if the features are scaled.

    4. Interpretable Coefficients in Regression : in order the coefficient of each feature represents the relative contribution of that feature to the prediction, Without scaling, the coefficients become difficult to interpret due to the different scales of the features.

    5. Consistency in Regularization : Ridge regression (L2 regularization) or Lasso regression (L1 regularization), are sensitive to feature scales. Without scaling, the regularization penalty can be uneven across features, leading to overfitting or underfitting.

**Disctinction Between Normalization and Standarization**

* Normalization: `Compresses data values into a specific range, such as [0, 1]`. It is used when we want to directly compare features with different scales.

* Standardization: Adjusts data so that `it has a mean of 0 and a standard deviation of 1`, ensuring the data follows a standard normal distribution.

**Normalization & Standarization Technique :**

* A. Normalization

    > main goal : convert the features into equal range

    1. Min-Max Scaling : $\frac {X-X_{min}}{X_{max} - X_{min}}$ : used while we have to maintain the distribution of data (features)

    2. MaxAbs Scaling : $\frac{X}{max|{X}|}$ : used while we have negative and positive data and we want to maintain the values it but with different range (-1, 1)

    3. Robust Scaler : $X_{norm} = \frac{X-Q_{1}}{Q_{3}-Q_{1}}$ : used while we face many outliers. scaling impact on outlier and rely on quartile values not min-max values. once we use this technique, the distribution of data would be changed

* B. Standarization

    > main goal : convert the features in order to have mean = 0 and standard deviation =1

    1. Z-Score Standarization : $X_{std} = \frac{X-\mu}{\sigma}$ : used while we have normal / gaussian distribution

    2. L1 Normalization (Least Absolute Normalization) : $X_{L1} = \frac{X}{|X_{1}| + |X_{2}| +.....+|X_{n}|}$ : Used for Regularization or you want to the output = 1.

    3. L2 Normalization (Least Squares) : $X_{L2} = \frac{X}{\sqrt{{X_{1}^2}+X_{2}^2+....+X_{n}^2}}$ : if you want the features distributed uniform and most useful in Ridge Regularization (L2 Regularization)

    4. BoxCox Normalization : $y(\lambda) =\begin{cases}\frac{y^\lambda - 1}{\lambda} & \text{if } \lambda \ne 0 \\\log(y) & \text{if } \lambda = 0\end{cases}$
    * The Box-Cox transformation is a statistical technique used to stabilize variance and make the data more normally distributed. It is particularly useful when dealing with data that exhibits non-constant variance (heteroscedasticity) or when the data is not normally distributed.

**Conclusion**

By looking at all the differences , `we can say that normal distribution should be the suitable choice when the data’s distribution do not match a gaussian distribution`. A practical transformation approach that helps your model perform and be more accurate is normalization.

Normalization of a machine learning model is helpful when you are unsure about the precise feature distribution. To put it another way, the data’s feature distribution does not have a Gaussian distribution.

Outliers in your data will be impacted by normalization because it needs a wide range to function correctly.

When you are entirely aware of the feature distribution of your data, or, to put it another way, when your data has a Gaussian distribution, standardization in the machine learning model is useful. This need not necessarily be the case, though. In contrast to Normalization, Standardization does not always have a bounding range; therefore, any outliers in your data won’t be impacted by it.

Scales for normalization fall between [0,1] and [-1,1]. Standardization has no range restrictions. When the algorithms don’t make any assumptions about the distribution of the data, Normalization is taken into account. When algorithms create predictions about the data distribution, standardization is applied.

## **A.Normalization Techniques**

### A.1 Min-Max Scaler

```python
# packages
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# dummy dataset
np.random.seed(0)
minmax_dummy_data = np.random.exponential(size=500).reshape(-1, 1)

# implement min max scaler
min_max_scaler_model = MinMaxScaler(feature_range=(0,1))
min_max_scaler_dataset = min_max_scaler_model.fit_transform(minmax_dummy_data)

# show the ten original and scaled data
print(f'raw data :\n{minmax_dummy_data[1:10]}\n\nMin-Max Scaled Data :\n {min_max_scaler_dataset[1:10]}')

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(8, 4))

# Plot the original data on the left subplot
sns.histplot(minmax_dummy_data, ax=ax[0], kde=True, alpha=0.2, legend=False)
ax[0].set_title('Original Data')

# Plot the scaled data on the right subplot
sns.histplot(min_max_scaler_dataset, ax=ax[1], kde=True, alpha=0.2, legend=False)
ax[1].set_title('Scaled Data')

# Show the plot
plt.tight_layout()
plt.show()


```

### **A.3 Robust Scaler**

```python
# import packages
from sklearn.preprocessing import RobustScaler
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# generate random sample
np.random.seed(0)

# normal dataset
robust_dummy_data = np.random.exponential(size=500)

# outlier dataset
robust_dummy_data_outlier = np.random.uniform(low=5, high=15, size=50)
robust_dummy_data[450:] = robust_dummy_data_outlier

# implement robust scaler
robust_scaler_model = RobustScaler(quantile_range=(25.0, 75.0), with_centering=True, with_scaling=True)
robust_scaled_data = robust_scaler_model.fit_transform(robust_dummy_data.reshape(-1, 1))


# visualize

# boxplot original data
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
sns.boxplot(robust_dummy_data, ax=ax[0,0], legend=False)
ax[0,0].set_title('Box Plot of Original Data')

# histplot original data
sns.histplot(robust_dummy_data, ax=ax[0,1], legend=False, kde=True, alpha = 0.5)
ax[0,1].set_title('Histogram of Original Data')

# boxplot robust scaled data
sns.boxplot(robust_scaled_data, ax=ax[1,0], legend=False)
ax[1,0].set_title('Box Plot of Scaled Data')

# histplot robust scaled data
sns.histplot(robust_scaled_data, ax=ax[1,1], legend=False, kde=True, alpha = 0.5)
ax[1,1].set_title('Histogram of Scaled Data')

plt.tight_layout()
plt.show()
```

## **B.Standarization Techniques**

### **B.1 Z-Score Standarization**

```python
# import packages
from sklearn.preprocessing import StandardScaler
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# dummy zscore dataset
np.random.seed(0)
dummy_data_zscore = np.random.random_sample((500,1))

# implement z-score standarization
zscore_standardscaler_model = StandardScaler(with_mean=True, with_std=True)
zscore_normalize_data = zscore_standardscaler_model.fit_transform(dummy_data_zscore.reshape(-1, 1))

# visualize dataset
fig, ax = plt.subplots(1, 2, figsize=(8, 6))

sns.histplot(dummy_data_zscore, kde=True, alpha=0.5, ax=ax[0], legend=False)
ax[0].set_title('Original Data')

sns.histplot(zscore_normalize_data, kde=True, alpha=0.5,ax=ax[1], legend=False)
ax[1].set_title('Normalized Data')

plt.tight_layout()
plt.show()


def print_mean_sd (dataset : np.array):
    import numpy as np
    mu = np.mean(dataset)
    std = np.std(dataset)
    
    print(f'mean : {mu} & std : {std}\n')

# mean and standard deviation of original dataset
print_mean_sd(dummy_data_zscore)
# mean and standard deviation of Normalized dataset
print_mean_sd(zscore_normalize_data)


```

### **B.3 L2 Normalization (Least Square Normalization or Euclidean Norm)**

```python
# import packages

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize

# generate random sample
np.random.seed(0)
l2_dummy_data = np.random.random_sample((500,2))+1

# implement l2 normalize
l2_normalize_data = normalize(l2_dummy_data, norm = 'l2', axis=1)

# visualize
fig, ax = plt.subplots(1,2,figsize=(8,6))
sns.kdeplot(l2_dummy_data, ax=ax[0])
ax[0].set_title('Original Data')

sns.kdeplot(l2_normalize_data, ax=ax[1])
ax[1].set_title('Normalized Data')

plt.tight_layout()
plt.show()

print_mean_sd(l2_dummy_data[:,0])
print_mean_sd(l2_normalize_data[:,0])
```

**What is Binarization ?**

Binarization is a data preprocessing technique used to convert data into binary form, where each value is either 0 or 1. This process is often applied to numerical or categorical data to simplify its representation and facilitate analysis or model training.

**Key Concepts:**

* Purpose:

    * Simplify Data: Binarization can simplify data, making it easier to handle in certain types of analyses or models.
    * Prepare for Machine Learning: In machine learning, especially in classification tasks, binarization can transform features to binary values, which is useful for models that require binary inputs.

* How It Works:

    * Thresholding: For numerical data, binarization typically involves applying a threshold. Values above the threshold are set to 1, while values below the threshold are set to 0.
    * Categorical Data: For categorical data, each category can be represented as a binary vector (one-hot encoding), where only one element is 1 and all others are 0.

* Types:

    * Threshold Binarization: Converts continuous data into binary based on a specified threshold. For example, if the threshold is 0.5, values greater than or equal to 0.5 are set to 1, and values below are set to 0.
    * One-Hot Encoding: Converts categorical data into binary vectors. Each category is represented by a vector where only one position is 1, and all others are 0.

**Applications :**

Feature Engineering: In machine learning, binarization can help convert features into binary format, making them compatible with certain algorithms.
Data Compression: Binarization can reduce the complexity of data by representing it in a simpler binary format.

**Conclusion**

> Binarization is a valuable technique in data preprocessing, especially when preparing data for classification algorithms or simplifying the representation of features.

```python
# packages
from sklearn.preprocessing import Binarizer
import numpy as np

# dummy dataset
np.random.seed(10)
dummy_data = np.random.randint(low=-10.9, high=10.8, size=(2,6))

# implement binarizer -> convert initial data to biner data (1,0) according to a threshold
binarizer = Binarizer(threshold=3)
dummy_data_binarizer = binarizer.transform(dummy_data)

print(f'raw data :\n {dummy_data}\n\nbinarized data :\n {dummy_data_binarizer}')
```
